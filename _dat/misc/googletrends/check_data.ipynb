```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ec=pd.read_csv("./ecommerce_region_W.csv.gz")
```

```python
ec
```

```python
dfs=[]
for d,df in ec.groupby(["keyword","region"]):
    dfs.append(df.sort_values("date").reset_index(drop=True))
```

```python
print(len(dfs))
```

```python

for i, df in enumerate(dfs):
    fig,ax = plt.subplots()    
    df.value.plot(ax=ax)
    if i>10:
        break
```
